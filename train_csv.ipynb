```python
df = df.append({"Survived": 1, 'Name': 'Atif', "Sex": 'male'}, ignore_index = True)
df
# Adding a value. Show in bottom
```

```python
df
```

```python
index2drop = df[df["Age"].isnull()==True].index
dfDrop = df.drop(index2drop, axis = 0)
dfDrop
```

```python
dfDrop = df.dropna(how='any' , subset=["Age"])
dfDrop
```

```python
#fillna()
dfFillna = df.fillna(value = {"Age" : 0})
dfFillna
```

```python
#sort_values()r
df.sort_values(by='Age', ascending = False, inplace = True)
df.head()
```

```python
#sort_index()
df.sort_index(axis=0, ascending = True, inplace = True)
df.head()
```

```python
#grouping data
group = df[['Sex', 'Age']].groupby('Sex').mean()
group.head()
```

```python
#Rearranging data

import numpy
pivot_table = pd.pivot_table(df, values=['Survived'],
                                columns=['Sex'],
                                 index=['Pclass'], 
                                 aggfunc={'Survived': numpy.sum})
pivot_table
```

```python
#Pandas DataFrames and Series can be plotted using the plot function,
#which uses the library for graphics Matplotlib.


pivot_table.plot(kind='bar', title="Survival by sex")


# <AxesSubplot:title={'center':'Survival by Sex'}, xlabel='Pclass'>
```
